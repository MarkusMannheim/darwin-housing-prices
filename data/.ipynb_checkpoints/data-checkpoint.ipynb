```python
import pandas as pd
import geopandas as gpd

data_prices = pd.read_excel("202109_NT Suburb data.xlsx",  header=1, usecols=[3,4,5,6], names=["name", "type", "value_2021", "value_2020"])

suburbs = gpd.read_file("SA2_2021_AUST_SHP_GDA2020")
suburbs = suburbs[
    (suburbs["SA3_CODE21"] == "70102") |
    (suburbs["SA3_CODE21"] == "70101") |
    (suburbs["SA3_CODE21"] == "70103") |
    (suburbs["SA3_CODE21"] == "70104")
]
suburbs = suburbs[["SA2_NAME21", "geometry"]]
suburbs.columns = ["name", "geometry"]
suburbs["class"] = "suburb"

mesh = gpd.read_file("MB_2021_AUST_SHP_GDA2020")
mesh = mesh[
    (mesh["MB_CAT21"] == "Water") & 
    (
        (mesh["SA3_CODE21"] == "70102") |
        (mesh["SA3_CODE21"] == "70103") |
        (mesh["SA3_CODE21"] == "70101") |
        (mesh["SA3_CODE21"] == "70104") |
        (mesh["SA3_CODE21"] == "70203")
    )
]
mesh = mesh[["geometry"]]
mesh["class"] = "water"

for i in suburbs.index:
    unit_data = data_prices[(data_prices["name"].apply(lambda x: x in suburbs.at[i, "name"])) & (data_prices["type"] == "Units")]
    house_data = data_prices[(data_prices["name"].apply(lambda x: x in suburbs.at[i, "name"])) & (data_prices["type"] == "Houses")]
    if len(unit_data) > 0:
        suburbs.at[i, "units_2021"] = unit_data.iat[0, 2]
        suburbs.at[i, "units_2020"] = unit_data.iat[0, 3]
    else:
        suburbs.at[i, "units_2021"] = None
        suburbs.at[i, "units_2020"] = None
    if len(house_data) > 0:
        suburbs.at[i, "houses_2021"] = house_data.iat[0, 2]
        suburbs.at[i, "houses_2020"] = house_data.iat[0, 3]
    else:
        suburbs.at[i, "houses_2021"] = None
        suburbs.at[i, "houses_2020"] = None

suburbs.plot(figsize=(15, 15));

mesh.to_file("./water.geojson", driver="GeoJSON")
suburbs.to_file("./suburbs.geojson", driver="GeoJSON")
```
